```python
import numpy as np
drug_data=np.load(r'after_merge\A375\drug_data.npy')#520*978
drug_name=np.load(r'after_merge\A375\drug_name.npy')#520*13 b'BRD-A17883755'

new_label=np.load(r'after_merge\A375\new_label.npy')#363*520

rngene_data=np.load(r'after_merge\A375\rngene_data.npy')#2754*978
rngene_data_sampled=np.load(r'after_merge\A375\rngene_data_sampled.npy')#2201*978
rngene_name=np.load(r'after_merge\A375\rngene_name.npy')#2754*5  b'G3BP1'
rngene_name_sampled=np.load(r'after_merge\A375\rngene_name_sampled.npy')#2201*5

tp_data=np.load(r'after_merge\A375\tp_data.npy')#363*978
tp_name=np.load(r'after_merge\A375\tp_name.npy')#363*6 b'PPP3R2'
```

```python
#处理正样本
pos_data=[]
count=0
for i in range(len(new_label)):
    for j in range(len(new_label[0])):
        if new_label[i][j]:
            pos_data.append([])
            pos_data[count].append(drug_data[j])
            pos_data[count].append(tp_data[i])
            count+=1    
```

```python
pos_data=np.array(pos_data)
```

```python
pos_data=pos_data.reshape(796,2*978)
```

```python
np.save(r'after_merge\A375\pos_data.npy',pos_data)
```

```python
#处理负样本
neg_data=[]
count=0
for i in range(len(drug_data)):
    for j in range(len(rngene_data_sampled)):
        neg_data.append([])
        neg_data[count].append(drug_data[i])
        neg_data[count].append(rngene_data_sampled[j])
        count+=1    
```
